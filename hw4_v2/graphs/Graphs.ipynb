```python
def plot_and_save(fname, mean, std, ind):
#     print mean, std, ind
    pl.figure(figsize=(7, 7))  # Don't create a humongous figure
    pl.errorbar(ind, mean, std, linestyle='None', color='r', marker='^')
    pl.xlabel("episode")
    pl.ylabel("undiscounted mean rewards")
    pl.savefig(fname, format='png') 
```

```python
# Enter Fourier basis order: 2
# Enter alpha: 0.001
# Enter lambda: 0.1
# Enter epsilon: 0.01
# Enter gamma: 1
# Enter number of trials to run: 100
# Enter the number of episodes per trial: 300
df = pd.read_csv("build/gcc/out_QLambda_MountainCar.csv")
print len(df)
plot_and_save("qlambda_car.png", df["Mean Return"], df["Standard Deviation"], range(len(df)))
```
